```python
# Q No.7. Write a program which takes 5 inputs from user for different subjectâ€™s marks, total it and generate mark sheet using grades ?
mark = []
tot = 0
print("Enter Marks Obtained in 5 Subjects: ")
for i in range(5):
    mark.insert(i, input())

for i in range(5):
    tot = tot + int(mark[i])
avg = tot/5

if avg>=80 and avg<=100:
    print("Your Grade is A+")
elif avg>=70 and avg<80:
    print("Your Grade is A")
elif avg>=60 and avg<70:
    print("Your Grade is B")
elif avg>=50 and avg<60:
    print("Your Grade is C")
elif avg>=33 and avg<50:
    print("Your Grade is E")
elif avg>=0 and avg<33:
    print("Your Grade is Fail")
else:
    print("Invalid Input!")
```

```python
# Q No.10.Write a Python program to sum all the numeric items in a list?
def sum_list(items):
    sum_numbers = 0
    for x in items:
        sum_numbers += x
    return sum_numbers
print(sum_list([1,2,-8, 8]))
```

```python
# Q No.11.Write a Python program to get the largest number from a numeric list.
def max_num_in_list( list ):
    max = list[ 0 ]
    for a in list:
        if a > max:
            max = a
    return max
print(max_num_in_list([10,20,30,40,50,-10,-20,-30]))
```
